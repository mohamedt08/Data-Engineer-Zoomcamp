```python
import pandas as pd
```

```python
prefix = 'https://github.com/DataTalksClub/nyc-tlc-data/releases/download/misc/'
url= f'{prefix}taxi_zone_lookup.csv'
taxi_zones_df = pd.read_csv(url)
taxi_zones_df
```

```python
df = pd.read_parquet('green_tripdata_2025-11.parquet')
df.head()
```

```python
from sqlalchemy import create_engine
engine = create_engine('postgresql://root:root@localhost:5432/ny_taxi')
```

```python
print(pd.io.sql.get_schema(taxi_zones_df, name='taxi_zones', con=engine))
```

```python
df.head(0).to_sql(name='green_taxi_data', con=engine, if_exists='replace')
taxi_zones_df.head(0).to_sql(name='taxi_zones_data', con=engine, if_exists='replace')
```

```python
taxi_zones_df.to_sql(name='taxi_zones_data', con=engine, if_exists='append')
df.to_sql(name='green_taxi_data', con=engine, if_exists='append')
```
